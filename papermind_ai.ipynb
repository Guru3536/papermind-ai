```python
import spacy
import scispacy

# en_core_sci_sm detects scientific term spans generally (a single 'ENTITY' label)
# rather than typed Method/Dataset/Metric labels, so we classify types ourselves
# below using curated gazetteers of common ML terms. Extend these lists freely --
# they're intentionally simple so you can grow them as you inspect real results.
nlp = spacy.load("en_core_sci_sm")
```

```python
KNOWN_DATASETS = {
    "mnist", "cifar-10", "cifar-100", "imagenet", "coco", "ms-coco", "squad", "glue",
    "wikitext", "imdb", "penn treebank", "librispeech", "cityscapes", "pascal voc",
    "kitti", "ade20k", "openai gym", "atari", "celeba", "lfw", "wmt", "conll", "snli",
    "multinli", "trec", "yelp", "ag news", "reuters", "common crawl", "wikipedia",
    "bookcorpus", "laion", "shapenet", "modelnet", "nyu depth", "kinetics", "ucf101",
    "hmdb51", "vggface", "visual genome", "vqa", "flickr30k", "charades"
}

KNOWN_METHODS = {
    "bert", "gpt", "gpt-2", "gpt-3", "roberta", "t5", "xlnet", "elmo", "word2vec",
    "glove", "fasttext", "resnet", "vgg", "alexnet", "inception", "densenet",
    "efficientnet", "mobilenet", "u-net", "yolo", "faster r-cnn", "mask r-cnn", "ssd",
    "gan", "vae", "lstm", "gru", "rnn", "cnn", "transformer", "attention",
    "self-attention", "autoencoder", "random forest", "svm", "xgboost", "lightgbm",
    "naive bayes", "k-means", "dbscan", "reinforcement learning", "q-learning",
    "deep q-network", "dqn", "actor-critic", "ppo", "a3c", "diffusion model",
    "stable diffusion", "clip", "vision transformer", "vit", "graph neural network",
    "gnn", "gcn", "capsule network"
}

KNOWN_METRICS = {
    "accuracy", "precision", "recall", "f1 score", "f1-score", "auc", "roc", "map",
    "bleu", "rouge", "perplexity", "iou", "mean squared error", "mse", "rmse", "mae",
    "r2 score", "log loss", "top-1 accuracy", "top-5 accuracy", "eer", "wer", "psnr", "ssim"
}

def classify_entity(text):
    t = text.lower().strip()
    if t in KNOWN_DATASETS:
        return "DATASET"
    if t in KNOWN_METHODS:
        return "METHOD"
    if t in KNOWN_METRICS:
        return "METRIC"
    return "CONCEPT"

def extract_entities(text):
    doc = nlp(text)
    return [{"text": ent.text, "type": classify_entity(ent.text)} for ent in doc.ents]

# quick check
extract_entities(df["paper_text"].iloc[0])
```

## 2. Corpus-wide caching for entities and keywords

Same pattern as your embeddings/FAISS caching: compute once, save to disk, reload instantly on future runs. This is what makes the hybrid search and topic map below actually usable -- without it, every feature would recompute NER/keywords at query time.

```python
import json
import os
from tqdm.auto import tqdm

# NOTE: NER + KeyBERT over all 15,000 papers can take a while on CPU.
# Set N_PAPERS to a smaller number (e.g. 3000) for a quick end-to-end test,
# then bump it up once you know the pipeline works.
N_PAPERS = len(df)

if os.path.exists("paper_entities.json"):
    print("Loading cached entities")
    with open("paper_entities.json") as f:
        all_entities = json.load(f)
else:
    print("Extracting entities for all papers")
    all_entities = []
    texts = df["paper_text"].tolist()[:N_PAPERS]
    for doc in tqdm(nlp.pipe(texts, batch_size=64), total=len(texts)):
        all_entities.append([{"text": ent.text, "type": classify_entity(ent.text)} for ent in doc.ents])
    with open("paper_entities.json", "w") as f:
        json.dump(all_entities, f)
    print("Entities cached to paper_entities.json")

df.loc[:len(all_entities)-1, "entities"] = pd.Series(all_entities, index=df.index[:len(all_entities)])
```

```python
if os.path.exists("paper_keywords.json"):
    print("Loading cached keywords")
    with open("paper_keywords.json") as f:
        all_keywords = json.load(f)
else:
    print("Extracting keywords for all papers")
    all_keywords = []
    for abstract in tqdm(df["abstract"].tolist()[:N_PAPERS]):
        kws = kw_model.extract_keywords(abstract, keyphrase_ngram_range=(1, 3), stop_words="english", top_n=8)
        all_keywords.append([k for k, _ in kws])
    with open("paper_keywords.json", "w") as f:
        json.dump(all_keywords, f)
    print("Keywords cached to paper_keywords.json")

df.loc[:len(all_keywords)-1, "keywords"] = pd.Series(all_keywords, index=df.index[:len(all_keywords)])
```

## 3. Hybrid search: semantic similarity + entity/keyword overlap, with explanations

Instead of ranking purely on cosine similarity, this re-ranks FAISS candidates using how many entities/keywords overlap with the query, and reports exactly which ones matched -- so results are explainable, not a black box.

```python
from collections import defaultdict

entity_to_papers = defaultdict(set)
for i, ents in enumerate(df["entities"].dropna()):
    for e in ents:
        entity_to_papers[e["text"].lower()].add(i)

print(f"Indexed {len(entity_to_papers)} unique entities across {df['entities'].notna().sum()} papers")
```

```python
def hybrid_search(query, top_k=5, candidate_k=50, alpha=0.7, entity_filter=None, summarize=False):
    """
    query: search text
    top_k: number of results to return
    candidate_k: candidates pulled from FAISS before re-ranking
    alpha: weight on semantic similarity vs entity/keyword overlap (0-1)
    entity_filter: optional list of entity/keyword strings a paper must contain
    summarize: if True, also generate an abstractive summary for each result
    """
    query_embedding = model.encode([query]).astype("float32")
    faiss.normalize_L2(query_embedding)
    D, I = index.search(query_embedding, candidate_k)

    query_entity_texts = set(e["text"].lower() for e in extract_entities(query))
    filter_set = set(x.lower() for x in entity_filter) if entity_filter else None

    results = []
    for sim_score, idx in zip(D[0], I[0]):
        row = df.iloc[idx]
        paper_entities = row["entities"] if isinstance(row["entities"], list) else []
        paper_keywords = row["keywords"] if isinstance(row["keywords"], list) else []
        paper_terms = set(e["text"].lower() for e in paper_entities) | set(k.lower() for k in paper_keywords)

        if filter_set and not (filter_set & paper_terms):
            continue

        overlap = query_entity_texts & paper_terms
        entity_score = len(overlap) / max(len(query_entity_texts), 1)
        final_score = alpha * float(sim_score) + (1 - alpha) * entity_score

        result = {
            "title": row["title"],
            "abstract": row["abstract"],
            "semantic_score": float(sim_score),
            "matched_terms": sorted(overlap),
            "final_score": final_score,
        }
        if summarize:
            result["summary"] = summarizer(row["abstract"], max_length=120, min_length=40, do_sample=False)[0]["summary_text"]
        results.append(result)

    return sorted(results, key=lambda r: r["final_score"], reverse=True)[:top_k]
```

```python
# Demo: hybrid search with explanation
for r in hybrid_search("deep learning for medical image analysis", summarize=True):
    print(f"{r['final_score']:.3f} | {r['title']}")
    if r["matched_terms"]:
        print(f"   matched: {r['matched_terms']}")
    print(f"   summary: {r['summary']}")
    print()
```

## 4. Topic map: unsupervised clustering of the whole corpus

Clusters papers by embedding similarity, auto-labels each cluster using its most common keyphrases, and plots a 2D map of the research landscape -- turning the project from a single-query search tool into something that shows the *shape* of the field.

```python
from sklearn.cluster import KMeans

N_CLUSTERS = 15
kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(embeddings)
df.loc[:len(cluster_labels)-1, "cluster"] = cluster_labels
```

```python
from collections import Counter

def label_cluster(cluster_id, top_n=5):
    papers = df[df["cluster"] == cluster_id]
    counter = Counter()
    for kws in papers["keywords"].dropna():
        counter.update(k.lower() for k in kws)
    return [term for term, _ in counter.most_common(top_n)]

cluster_names = {c: ", ".join(label_cluster(c)) for c in range(N_CLUSTERS)}
for c, name in cluster_names.items():
    print(f"Cluster {c}: {name}")
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

coords = PCA(n_components=2, random_state=42).fit_transform(embeddings)
df.loc[:len(coords)-1, "x"] = coords[:, 0]
df.loc[:len(coords)-1, "y"] = coords[:, 1]

plt.figure(figsize=(12, 8))
scatter = plt.scatter(df["x"], df["y"], c=df["cluster"], cmap="tab20", s=6, alpha=0.6)

for c in range(N_CLUSTERS):
    cx, cy = df[df["cluster"] == c][["x", "y"]].mean()
    plt.text(cx, cy, cluster_names[c], fontsize=8, weight="bold",
             ha="center", bbox=dict(facecolor="white", alpha=0.7, edgecolor="none"))

plt.title("Topic Map of ML ArXiv Papers")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.tight_layout()
plt.show()
```

```python
import gradio as gr

def gradio_search(query, filter_term, do_summarize):
    entity_filter = [filter_term.strip()] if filter_term.strip() else None
    results = hybrid_search(query, top_k=5, entity_filter=entity_filter, summarize=do_summarize)

    if not results:
        return "No results matched your filter. Try removing it or using a different term."

    output = ""
    for r in results:
        output += f"### {r['title']}\n"
        output += f"**Score:** {r['final_score']:.3f}"
        if r["matched_terms"]:
            output += f" | **Matched:** {', '.join(r['matched_terms'])}"
        output += "\n\n"
        if do_summarize:
            output += f"**Summary:** {r['summary']}\n\n"
        else:
            output += f"{r['abstract'][:400]}...\n\n"
        output += "---\n"
    return output

demo = gr.Interface(
    fn=gradio_search,
    inputs=[
        gr.Textbox(label="Search query", placeholder="e.g. deep learning for medical image analysis"),
        gr.Textbox(label="Filter by method/dataset (optional)", placeholder="e.g. cnn"),
        gr.Checkbox(label="Generate summaries", value=True),
    ],
    outputs=gr.Markdown(),
    title="Semantic Paper Search",
    description="Search ML ArXiv papers by meaning, filter by method/dataset, and get auto-generated summaries.",
)

demo.launch()
```

# Upgrade: Research Gap Finder & Real Trend Analysis

Two features that push this from "a search tool" toward "a tool that produces insight":
1. **Research gap finder** — uses the Method/Dataset entities already extracted to find combinations that are common individually but rarely or never tried together
2. **Trend analysis** — pulls real publication dates from the arXiv API to show how method usage has actually changed over time (not guessed, measured)

## 1. Research gap finder

This only uses data you already have (`df["entities"]`), no network calls, so it's fast and reliable. For each pair of (top method, top dataset), it compares how often they actually co-occur in a paper against how often you'd *expect* them to co-occur if method choice and dataset choice were independent. Pairs with high expected-but-zero-actual co-occurrence are candidates for "nobody has tried this combination yet."

**Caveat to be upfront about:** this is a co-occurrence heuristic, not a claim about the real literature — a zero here means "not in this 15k-paper sample," not "never done by anyone, ever." Frame it in your writeup as a hypothesis-generation tool, not a ground truth.

```python
import itertools
import numpy as np
from collections import defaultdict, Counter

TOP_N = 15
top_methods = [m for m, _ in top_entities_by_type("METHOD", TOP_N)]
top_datasets = [d for d, _ in top_entities_by_type("DATASET", TOP_N)]

co_occurrence = defaultdict(int)
method_totals = Counter()
dataset_totals = Counter()

for ents in df["entities"].dropna():
    methods_in_paper = set(e["text"].lower() for e in ents if e["type"] == "METHOD" and e["text"].lower() in top_methods)
    datasets_in_paper = set(e["text"].lower() for e in ents if e["type"] == "DATASET" and e["text"].lower() in top_datasets)

    for m in methods_in_paper:
        method_totals[m] += 1
    for d in datasets_in_paper:
        dataset_totals[d] += 1
    for m, d in itertools.product(methods_in_paper, datasets_in_paper):
        co_occurrence[(m, d)] += 1

total_papers_with_entities = df["entities"].notna().sum()
print(f"Built co-occurrence stats from {total_papers_with_entities} papers")
```

```python
# Heatmap: darker = more papers pairing that method with that dataset
matrix = np.zeros((len(top_methods), len(top_datasets)))
for i, m in enumerate(top_methods):
    for j, d in enumerate(top_datasets):
        matrix[i, j] = co_occurrence.get((m, d), 0)

plt.figure(figsize=(12, 10))
plt.imshow(matrix, cmap="YlOrRd")
plt.xticks(range(len(top_datasets)), top_datasets, rotation=90)
plt.yticks(range(len(top_methods)), top_methods)
plt.colorbar(label="papers mentioning both")
plt.title("Method x Dataset co-occurrence")
plt.tight_layout()
plt.show()
```

```python
# Flag combinations that are individually common but essentially never paired together
gaps = []
for m in top_methods:
    for d in top_datasets:
        actual = co_occurrence.get((m, d), 0)
        expected = (method_totals[m] * dataset_totals[d]) / max(total_papers_with_entities, 1)
        if expected > 1 and actual == 0:
            gaps.append((m, d, expected))

gaps.sort(key=lambda x: -x[2])

print("Potential under-explored (method, dataset) combinations in this corpus:\n")
for m, d, exp in gaps[:15]:
    print(f"  {m}  x  {d}   (expected ~{exp:.1f} papers if independent, found: 0)")
```

```python
import feedparser
import urllib.request
import urllib.parse
import time
import json
import os

N_TREND_SAMPLE = 15  # increase once you've confirmed this works; each lookup takes ~3s

def fetch_arxiv_date(title):
    query = urllib.parse.quote(f'ti:"{title}"')
    url = f"http://export.arxiv.org/api/query?search_query={query}&max_results=1"
    try:
        with urllib.request.urlopen(url, timeout=10) as resp:
            feed = feedparser.parse(resp.read())
        if feed.entries:
            return feed.entries[0].published[:10]  # YYYY-MM-DD
    except Exception:
        pass
    return None

if os.path.exists("paper_dates.json"):
    print("Loading cached dates")
    with open("paper_dates.json") as f:
        paper_dates = json.load(f)
else:
    print(f"Fetching publication dates for {N_TREND_SAMPLE} papers from arXiv (this will take a few minutes)")
    paper_dates = {}
    sample_df = df.head(N_TREND_SAMPLE)
    for i, title in enumerate(sample_df["title"]):
        paper_dates[str(sample_df.index[i])] = fetch_arxiv_date(title)
        time.sleep(3)  # respect arXiv's rate limit guidance
    with open("paper_dates.json", "w") as f:
        json.dump(paper_dates, f)
    print("Dates cached to paper_dates.json")

found = sum(1 for v in paper_dates.values() if v)
print(f"Found dates for {found}/{len(paper_dates)} sampled papers")
```

# Upgrade: RAG-based Q&A

Everything so far retrieves *papers*. This adds a layer that answers *questions*, synthesizing an answer from multiple retrieved abstracts with citations back to the specific papers each claim came from -- the core idea behind retrieval-augmented generation.

**This is a genuinely different capability from summarization.** Summarization condenses one paper. RAG here reads several papers at once and answers a question that no single paper's abstract directly answers on its own, e.g. "what approaches have been used for X" pulled from five different papers.

## 1. Load a generation model

Using `google/flan-t5-base` here since it's small enough to run on CPU and instruction-tuned, so it follows a "use only the sources below" prompt reasonably well. If you have GPU/RAM to spare, swap in `google/flan-t5-large` or `google/flan-t5-xl` for noticeably better answers -- the RAG logic below doesn't change, only the model name.

```python
from transformers import pipeline

rag_generator = pipeline(
    "text2text-generation",
    model="google/flan-t5-base"
)
```

## 2. The RAG pipeline: retrieve, build a grounded prompt, generate, cite

Steps:
1. Retrieve top-k candidates with your existing `hybrid_search`
2. Truncate each abstract so the combined context fits the model's input window
3. Prompt the model to answer **only** using the numbered sources, and to say it doesn't know if the sources don't cover the question
4. Return the answer alongside the actual paper titles behind each `[n]` reference, so the citation is checkable, not just a claim

```python
def build_rag_prompt(question, sources, max_chars_per_source=350):
    context_blocks = []
    for i, s in enumerate(sources, start=1):
        snippet = s["abstract"][:max_chars_per_source]
        context_blocks.append(f"[{i}] {snippet}")

    context = "\n\n".join(context_blocks)
    prompt = (
        "Answer the question using ONLY the numbered sources below. "
        "Reference sources by their number, like [1] or [2]. "
        "If the sources do not contain enough information to answer, say so clearly.\n\n"
        f"Sources:\n{context}\n\n"
        f"Question: {question}\n"
        "Answer:"
    )
    return prompt


def rag_answer(question, k=5):
    sources = hybrid_search(question, top_k=k)

    if not sources:
        return {"answer": "No relevant papers were found for this question.", "sources": []}

    prompt = build_rag_prompt(question, sources)
    generated = rag_generator(prompt, max_length=200, do_sample=False)[0]["generated_text"]

    return {
        "answer": generated,
        "sources": [{"ref": i + 1, "title": s["title"]} for i, s in enumerate(sources)]
    }
```

```python
# Demo
result = rag_answer("What approaches have been used to handle class imbalance in medical imaging?")

print("Answer:")
print(result["answer"])
print("\nSources:")
for s in result["sources"]:
    print(f"  [{s['ref']}] {s['title']}")
```

## 3. Honest limitations (worth stating explicitly in your writeup)

- **flan-t5-base is a small model.** It follows the "cite your sources" instruction reasonably often but not perfectly -- it can still merge facts across sources incorrectly, or cite the wrong number. Manually check a handful of answers against the actual abstracts before presenting this as reliable.
- **Grounding reduces but doesn't eliminate hallucination.** The model is instructed to use only the provided sources, but nothing prevents it from adding a plausible-sounding but unsupported detail. This is a real limitation of small-model RAG, not a bug in this code.
- **Answer quality scales with the retrieval quality.** If `hybrid_search` returns weakly relevant papers for an unusual question, the answer will be weak too -- RAG can't answer questions the corpus doesn't cover.
- **Truncating abstracts to 350 characters** keeps prompts within the model's context window, but can cut off the specific detail the question is asking about. Worth trying summaries instead of raw truncation as a future improvement (pass through `summarizer` first).

## 4. Add Q&A as a second tab in the interactive demo

Upgrades the earlier single-purpose Gradio app into a two-tab interface: paper search, and question answering over the corpus.

```python
import gradio as gr

def gradio_search(query, filter_term, do_summarize):
    entity_filter = [filter_term.strip()] if filter_term.strip() else None
    results = hybrid_search(query, top_k=5, entity_filter=entity_filter, summarize=do_summarize)
    if not results:
        return "No results matched your filter. Try removing it or using a different term."
    output = ""
    for r in results:
        output += f"### {r['title']}\n"
        output += f"**Score:** {r['final_score']:.3f}"
        if r["matched_terms"]:
            output += f" | **Matched:** {', '.join(r['matched_terms'])}"
        output += "\n\n"
        output += f"**Summary:** {r['summary']}\n\n" if do_summarize else f"{r['abstract'][:400]}...\n\n"
        output += "---\n"
    return output


def gradio_ask(question, k):
    result = rag_answer(question, k=int(k))
    output = f"### Answer\n{result['answer']}\n\n### Sources\n"
    for s in result["sources"]:
        output += f"- [{s['ref']}] {s['title']}\n"
    return output


with gr.Blocks(title="Semantic Paper Search + RAG Q&A") as demo:
    gr.Markdown("# Semantic Paper Search + RAG Q&A")
    with gr.Tab("Search papers"):
        query_in = gr.Textbox(label="Search query", placeholder="e.g. deep learning for medical image analysis")
        filter_in = gr.Textbox(label="Filter by method/dataset (optional)", placeholder="e.g. cnn")
        summarize_in = gr.Checkbox(label="Generate summaries", value=True)
        search_btn = gr.Button("Search")
        search_out = gr.Markdown()
        search_btn.click(gradio_search, [query_in, filter_in, summarize_in], search_out)
    with gr.Tab("Ask a question"):
        question_in = gr.Textbox(label="Question", placeholder="e.g. what approaches handle class imbalance in medical imaging?")
        k_in = gr.Slider(minimum=3, maximum=8, value=5, step=1, label="Number of sources to use")
        ask_btn = gr.Button("Ask")
        ask_out = gr.Markdown()
        ask_btn.click(gradio_ask, [question_in, k_in], ask_out)

demo.launch()
```
